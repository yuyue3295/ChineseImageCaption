```python
import os
import sys
import pprint
import codecs
import jieba


input_description_file = "./data/clean_zh_results_20130124.token"
output_vocab_file = "./data/zh_vocab.txt"


def count_vocab(input_token_filename):
    with codecs.open(input_token_filename, 'r','utf-8') as f:
        lines = f.readlines()
    
    max_length_of_sentences = 0
    length_dict = {}
    vocab_dict = {}
    
    for line in lines:
        image_id, description = line.strip('\n').split('\t')
        '''
        1000092795.jpg#0	毛茸茸的头发的两个年轻人看着他们的手，同时在院子里闲逛。
        分成了两部分，图片名称和句子描述编号为一部分，具体的描述分为一部分。
        '''
        description = description.strip('\n')
        description = description.strip('\r\n')
        description = description.strip()
        if(len(list(description))<=3):
            print(image_id,'这个句子的长度是小于等于3的')
        words = jieba.cut(description)
#         words = description.strip(' ').split()#描述进行分词
        words = list(words)
        max_length_of_sentences = max(max_length_of_sentences, len(words))
        length_dict.setdefault(len(words), 0)#统计单词长度的出现的次数
        length_dict[len(words)] += 1
        #统计句子长度的字典，key是句子长度，value是这种长度的句子出现了多少次。

        for word in words:
            vocab_dict.setdefault(word, 0)
            vocab_dict[word] += 1 #统计单词出现的次数
    
    print(max_length_of_sentences)
    pprint.pprint(length_dict)
    return vocab_dict

vocab_dict = count_vocab(input_description_file)
```

```python
pprint.pprint(vocab_dict)
```

```python
sorted_vocab_dict = sorted(vocab_dict.items(), key = lambda d:d[1], reverse=True)
'''reverse = True 降序 ， reverse = False 升序（默认）'''
with codecs.open(output_vocab_file, 'w','utf-8') as f:
    f.write("<UNK>\t100000\n")
    for item in sorted_vocab_dict:
        f.write('%s\t%d\n' % item)
```

```python
import re
'''添加句号'''
with codecs.open("./data/zh_results_20130124.token", 'r','utf-8') as f:
        lines = f.readlines()
        with codecs.open('./data/clean_zh_results_20130124.token','w','utf-8') as g:
            for line in lines:
                line = line.strip('\n')
                re.sub(r' ',line,'')
                match = re.search(r'。',line)
                if match == None:
                    line = line+'。'
                line = line + '\n'
                g.write(line)
```
